# Zomato Data Analysis Project

## STEP 1- Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## STEP 2- Create the data frame

```python
dataframe= pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

## Convert the data type of column- rate

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
    
```

```python
dataframe.info()
```

# 1. Type of Restaurant that has max orders

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of restaurant")
```

## Conclusion 1- majority of the customer fall in dining category

# 2. Type of restaurant which has most no of votes

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of restaurant", c="red",size=20)
plt.ylabel("Votes",c="red",size=20)
```

## Conclusion 2- Dining category has got the most votes

# 3. Rating received by most restaurants

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("ratings distribution")
plt.show()
```

## Conclusion 3- the majority restaurants receive ratings from 3.5 to 4

# 4. Average order Spent by Couples

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

## Conclusion 4- the majority of couples prefer restaurants with an approx cost of 300 for two

# 5. Which mode receives maximum rating

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate', data=dataframe)
```

## Conclusion 5- people rate online orders more than offline orders

# 6. Which Type of Restaurant Receive Which Type of Orders

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In(Type)")
plt.show()
```

## Conclusion 6- Dining Restaurants primarily accept offline orders whereas cafes primarily receive online orders. This suggests that clients preferred  orders in person at restaurants, but prefer online ordering at cafes.
